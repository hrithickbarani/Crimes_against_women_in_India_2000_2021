```python
#importing libraries

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("darkgrid")
```

```python
#Read input data

df=pd.read_csv('CrimesOnWomenData.csv', usecols=['State', 'Year', 'Rape', 'K&A', 'DD', 'AoW', 'AoM', 'DV',
       'WT'])
```

```python
#Columns of df

df.columns
```

```python
df.columns= df.columns.str.lower()
```

```python
#Check values in state column

df['state'].unique()
```

```python
#Uniform names for state columns

df['state']=df['state'].str.upper()
```

```python
df['state']=df['state'].replace('D & N HAVELI','D&N HAVELI')
```

```python
df['state'].unique()
```

```python
#Renaming column names

df.rename(columns={'k&a':'kidnap_and_assault','dd':'dowry_deaths','aow':'physical_assault','aom':'assault_on_modesty','dv':'domestic_violence','wt':'women_trafficking'}, inplace=True)
```

```python
#Total crimes column, addition of all crimes

df['total_crimes']=df['rape']+df['kidnap_and_assault']+df['dowry_deaths']+df['physical_assault']+df['assault_on_modesty']+df['domestic_violence']+df['women_trafficking']
```

```python
#Removing two columns due to inconsistent data (Caught while plotting trends over the years)

df.drop(columns=['assault_on_modesty','women_trafficking'], inplace=True)
```

```python
#Crime-wise, which states topped most no. of times from 2001-2021

state_with_max_crimes=df.loc[df.groupby('year')['total_crimes'].idxmax()].reset_index(drop=True)
state_with_max_crimes_count=state_with_max_crimes['state'].value_counts().reset_index()

state_with_max_rapes=df.loc[df.groupby('year')['rape'].idxmax()].reset_index(drop=True)
state_with_max_rapes_count=state_with_max_rapes['state'].value_counts().reset_index()

state_with_max_domestic_violence=df.loc[df.groupby('year')['domestic_violence'].idxmax()].reset_index(drop=True)
state_with_max_domestic_violence_count=state_with_max_domestic_violence['state'].value_counts().reset_index()

state_with_max_kidnap_and_assault=df.loc[df.groupby('year')['kidnap_and_assault'].idxmax()].reset_index(drop=True)
state_with_max_kidnap_and_assault_count=state_with_max_kidnap_and_assault['state'].value_counts().reset_index()

state_with_max_dowry_deaths=df.loc[df.groupby('year')['dowry_deaths'].idxmax()].reset_index(drop=True)
state_with_max_dowry_deaths_count=state_with_max_dowry_deaths['state'].value_counts().reset_index()

state_with_max_physical_assault=df.loc[df.groupby('year')['physical_assault'].idxmax()].reset_index(drop=True)
state_with_max_physical_assault_count=state_with_max_physical_assault['state'].value_counts().reset_index()
```

```python
#Every year, Which state records highest number of rape cases?

fig, ax=plt.subplots(1,2,figsize=(13,5))
state_with_max_rapes.plot(kind='bar',x='year',y='rape',ax=ax[0],legend=False)
state_with_max_rapes_count[:3].plot(kind='bar',x='state',ax=ax[1])

for index,row in state_with_max_rapes.iterrows():
    ax[0].text(index,row['rape']+0.05*row['rape'], row['state'] ,rotation=90)

for bars in ax[1].containers:
    ax[1].bar_label(bars, fontsize=12)
    
ax[1].tick_params(rotation=0)

fig.suptitle('States with highest Rape cases, per year from 2001-2021', fontsize=17,fontweight='bold')
plt.tight_layout()
plt.show()
```

```python
years=[2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020,2021]


#Every year, which crime records the highest cases in india from 2001-2021

results=[]
for i in years:
    data=df[df['year']==i][['rape', 'kidnap_and_assault', 'dowry_deaths',
       'physical_assault', 'domestic_violence']]
    crimes_sum=data.sum()
    max_crime=crimes_sum.idxmax()
    max_cases=crimes_sum.max()

    results.append({'year':i, 'max_crime':max_crime, 'max_cases':max_cases})
    
max_crime_yearwise=pd.DataFrame(results)
max_crime_yearwise
```

```python
#Most recorded crime against women in 21 years, from 2001-2021

sum_of_crimes_list=[]

crimes=['rape', 'kidnap_and_assault', 'dowry_deaths',
       'physical_assault', 'domestic_violence']
for crime in crimes:
    sum=df[crime].sum()
    sum_of_crimes_list.append({'crime':crime,'sum_of_cases':sum})
    

sum_of_crimes_df=pd.DataFrame(sum_of_crimes_list)
sum_of_crimes_df=sum_of_crimes_df.sort_values(by='sum_of_cases',ascending=False).reset_index(drop=True)

ax=sum_of_crimes_df.plot(kind='bar',x='crime',y='sum_of_cases', figsize=(16,5))

for bars in ax.containers:
    ax.bar_label(bars, fmt='%.0f')
plt.xticks(rotation=0)
plt.title('Most recorded crime against women in india from 2001-2021', fontsize=15, fontweight='bold')
plt.show()
```

```python
#Top 5 States with most no. of crimes against women from 2001-2021

ax=df.groupby('state')[['total_crimes']].sum().sort_values(by='total_crimes',ascending=False)[:5].plot(kind='bar',figsize=(16,5))

#Bar label
for bars in ax.containers:
    ax.bar_label(bars)
        
#x-axis label rotation
ax.tick_params(axis='x',rotation=0)

ax.set_title('Top 5 states with highest crimes against women in india from 2001-2021', fontsize=15, fontweight='bold')

plt.show() 
```

```python
#Every year from 2001-2021, which state has most number of crimes against women from 2001-2021

state_with_max_crimes_yearwise= df.loc[df.groupby('year')['total_crimes'].idxmax(),['year', 'state', 'total_crimes'] ].reset_index(drop=True)
plt.figure(figsize=(18,5))
sns.barplot(x='year', y='total_crimes', hue='state', dodge=False, data=state_with_max_crimes_yearwise,palette='Set1')
plt.title('States with highest number of crimes against women, per year from 2001-2021', fontsize=15, fontweight='bold')
plt.show()
```

```python
#Corelation between crimes

crimes_for_corelation=df[['rape', 'kidnap_and_assault', 'dowry_deaths',
       'physical_assault', 'domestic_violence']].corr()

crimes_for_corelation
```

### 1) UTTAR PRADESH has the highest no. of Dowry death cases from 2001-2021, followed by *Bihar* and *Madhya Pradesh*.
### 2) In the year 2014, India recorded highest no. of Dowry death cases with 10,050 cases.

### 4) Dowry death cases has nor increased nor decreased from 2001-2021, it has a steady line of cases coming each year of around 8,000 cases.

## PHYSICAL ASSAULT:
### 1) MADHYA PRADESH has the highest no. of Physical assault on women cases, followed by *Maharashtra* and *Uttar pradesh*.
### 2) In 2018, India recorded highest no. of Physical assault on women cases with 89,097 cases.

### 4) Physical assault cases has been increasing steadily since 2001, Witnessed sudden spike in cases from 30,000 cases in 2012 to 70,000 cases in 2013 and upwards onwards.

### 2) In 2021, India recorded highest no. of Domestic violence cases with 1,29,139 cases.
